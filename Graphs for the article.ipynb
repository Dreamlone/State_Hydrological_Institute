```python
# Функция нормирования шкалы массива на диапазон от 0 до 1,75*Пи
def convert_to_polar(arr):
    # Нормируем на диапазон от 0 до 1
    arr = (arr/100) + 0.016
    # А теперь на полярные координаты
    arr = arr * (1.94*np.pi)
    return(arr)

convert_to_polar(np.array([5, 10, 15, 28, 44, 50, 74, 93]))
```

## Линейные графики с амплитудой

```python
d2 = pd.read_excel('D:/2019-2021 Магистратура ИТМО/ГГИ/MAE_gapfilling.xlsx')

for city in ['Spb', 'Madrid']:
    data_2 = d2[d2['Territory'] == city]      
    
    # Агрегируем данные
    df = data_2.groupby(['Territory', 'Product', 'Gap']).agg({'MAE': 'mean'}).reset_index()
    
    # Разделяем датасеты по продуктам
    df_sent = df[df['Product'] == 'Sentinel-3']
    data_2_sent = data_2[data_2['Product'] == 'Sentinel-3']
    data_2_sent = data_2_sent.groupby('Gap').agg({'R': 'mean'}).reset_index()
    
    df_mod1 = df[df['Product'] == 'MOD11A1']
    data_2_mod1 = data_2[data_2['Product'] == 'MOD11A1']
    data_2_mod1 = data_2_mod1.groupby('Gap').agg({'R': 'mean'}).reset_index()
    
    df_mod2 = df[df['Product'] == 'MOD11_L2']
    data_2_mod2 = data_2[data_2['Product'] == 'MOD11_L2']
    data_2_mod2 = data_2_mod2.groupby('Gap').agg({'R': 'mean'}).reset_index()
    
    # Отрисовка графика
    fig, ax1 = plt.subplots()    
    ax1.set_xlabel('Gap size, %', fontsize = 14)
    ax1.set_ylabel('MAE, ℃', fontsize = 14)
    ax1.plot(df_sent['Gap'], df_sent['MAE'], '-ok', color='blue', label = 'Sentinel-3')
    ax1.plot(df_mod1['Gap'], df_mod1['MAE'], '-ok', color='orange', label = 'MOD11A1')
    ax1.plot(df_mod2['Gap'], df_mod2['MAE'], '-ok', color='green', label = 'MOD11_L2')
    ax1.tick_params(axis='y')
    plt.grid(c='#DCDCDC')
    plt.legend(loc = 'upper left')
    
    ax2 = ax1.twinx()  
    ax2.plot(data_2_sent['Gap'], data_2_sent['R'], color='red', linewidth=1, 
             alpha = 0.5)
    ax2.fill_between(data_2_sent['Gap'], 0, data_2_sent['R'], color = 'red', alpha = 0.05, label = 'Temperature amplitude')
    ax2.set_ylabel('Temperature amplitude in the gap, ℃', fontsize = 13)
    ax2.tick_params(axis='y')
    plt.legend(loc = 'lower right')
    plt.title(city, fontsize = 15)
    plt.show()
```
